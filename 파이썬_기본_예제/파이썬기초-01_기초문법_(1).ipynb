````python
'''
  다음과 같이 함수를 정의하여 사용할 수 있습니다.
'''
# 함수의 정의
def int_add(A:int)-> str: # 함수를 정의 내리는 def 코드로 function 이라는 이름의 함수를 정의 인자로는 A를 답음
  B:int = int(input("B = ")) # 미리 정의되어 있는 함수도 함수 정의 내에 사용할 수 있음 input 함수는 사용자로부터 입력을 받는 함수, 키보드로 입력받은 문자를 정수형으로 케스팅하고 있다.
  C:int = A + B
  print("A + B = {}".format(C))
  return str(C) # 식별자가 가르키는 데이터 타입을 변경하는 것을 케스팅이라고 한다. str(C) 에서 str()을 사용하여 정수형인 C를 문자열로 바꾸고 있다.

# 정의한 함수를 실행해 본다.
int_add(4)
````

````python
# fucntion argument & Parameter Types

def complicated_function_1(x,y,z=None) -> None:
    print(x,y,z)
    pass

complicated_function_1(1,2,3)
complicated_function_1(3, y=2, z=1)

# complicated_function(x=1,2,z=3) error
complicated_function_1(1,2)


def complicated_function_2(x,y, *args) -> None:
    print(x,y,args)
    pass

complicated_function_2(1,2,4,5,7,8)

def complicated_function_3(*args, **kwargs) -> None:
    print(args, kwargs)
    pass

complicated_function_3(1,2,3,b=1, c=True, d="foo")

def complicated_function_4(a, b, c=True, d=False) -> None:
    print(a, b, c, d)
    pass

complicated_function_4(*[1,2],**{"c":"foo","d":"bar"})
````

````python
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])

s3 = s1 & s2 # s1.intersection(s2)

s4 = s1 | s2 # s1.union(s2)

s1 - s2 # s1.difference(s2)
s2 - s1 # s2.difference(s1)

s1.add(4)

s5 = set([1, 2, 3])
s5.update([4, 5, 6])

s5.remove(4)
````

|-----|-----|-----|

|  &  | 교집합|intersection()|
|  -  | 차집합|difference()|

# pip

- 파이썬 라이브러리 관리 프로그램

````python
python –c ‘import urlib’
#: cmd 터미널 상태에서 python 인터프리터 터미널로 들어가지 않고 명령 사용하여 모듈을 가져와 보기.

pip list
#: 설치된 모든 모듈을 표시함.

python show [모듈명]
#: 특정 설치된 모듈의 정보를 표시

pip freeze
#: 현재 설치된 리스트 표시

pip freeze > requirements.txt
#: 설치된 모듈 리스트를 파일로 저장하기.

pip install –r requirements.txt
#: 모듈 리스트 파일로부터 모듈 설치

pip uninstall [모듈명]
#: 모듈 삭제

pip install [모듈명]==version (부등호로 version 설정)
#: 모듈의 특정 버전 설치 조건 부여

pip download $PATH –r requirements.txt
#: 모듈 다운로드 하기

pip install [모듈명]
#: 모듈 설치
pip install —no-lndex —find-links=[경로]
#: 다운로드 한 모듈 설치
````
