### 2. 파이썬 기본 자료형

- 정수형 int
- 유리수 float
- 무자열 string

```python
# int 형
A = 3
B = 7
print(A)
print(B)

# float 형
C = 1.2
D = 1.5
print(C)
print(D)

# String 형
E = '안녕하세요 파이썬'
F = '반갑습니다.'
print(E)
print(F)
```

```python
# 자료형을 확인하느법
print(type(A))
print(type(C))
print(type(E))
```
